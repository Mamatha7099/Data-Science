# Principle Component Analysis

```python
from sklearn.decomposition import PCA  #importing liabraries
pca=PCA()
fit=pca.fit(X[numerical])  #selecting pca
```

```python
import matplotlib.pyplot as plt  #importing liabraries
import numpy as np
plt.figure.figsize=(5,5)  # giving plot size

PC_values = np.arange(pca.n_components_) + 1  # giving PC values
plt.plot(PC_values, pca.explained_variance_ratio_, 'o-', linewidth=2, color='blue')

plt.title('Scree Plot of a Elbow Plot') #giving title
plt.xlabel('Principal Component') #giving x AND y axis names
plt.ylabel('Variance Explained')
plt.show()
```

```python
per_var = np.round(pca.explained_variance_ratio_* 100, decimals=1) #showing pca values as plot
labels = ['PC' + str(x) for x in range(1, len(per_var)+1)]
plt.bar(x=range(1,len(per_var)+1), height=per_var, tick_label=labels)
plt.ylabel('Percentage of Explained Variance')#giving x and y label names 
plt.xlabel('Principal Component')
plt.title('Scree Plot')
plt.show()
```
